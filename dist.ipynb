```python
#poisson 

#  what is the probability of number of schools in range of 0 to 5 where the mean number of school in a 10km radius
## is 3.

from scipy.stats import poisson
import matplotlib.pyplot as plt

X = [1,2,3,4,5]
sample = poisson.pmf(3,X)
plt.plot(sample)
plt.ylabel("Probability")
plt.show()
```

```python
# normal dist--create a probability dist for a sample that has 100 random variables with mean 5 and sd 50

from scipy.stats import norm
import numpy as np
mean = 5
sd =50
sample = np.linspace(-100,100,1000)
f= norm(mean,sd)
dist = f.pdf(sample)
plt.plot(dist)
```
